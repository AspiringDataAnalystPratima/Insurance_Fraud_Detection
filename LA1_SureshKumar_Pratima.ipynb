```python
# en_core_web_lg is to a specific pre-trained spaCy model for the English language with a large vocabulary and word vectors.The lg in en_core_web_lg stands for "large," indicating that it is a larger and more comprehensive model compared to the smaller versions available.
import spacy.cli
spacy.cli.download("en_core_web_lg")
nlp = spacy.load("en_core_web_lg")
```

```python
#Converting review as a dataframe and combining one star and five star review into single dataframe
restaurant_reviews = r'/gdrive/My Drive/restaurant_reviews_az.csv'
restaurant_reviews= pd.read_csv(restaurant_reviews)
# Selecting the 1 star and 5 star reviews from the dataset
one_star_reviews = restaurant_reviews[restaurant_reviews['stars'] == 1]
five_star_reviews = restaurant_reviews[restaurant_reviews['stars'] == 5]

# Combine the 1 star and 5 star reviews into a single DataFrame
combined_reviews = pd.concat([one_star_reviews, five_star_reviews])

# Reset index of the combined DataFrame
combined_reviews.reset_index(drop=True, inplace=True)

combined_reviews.head()  # Display the first few rows of the filtered dataframe

```

```python
#Process of  tokenization of each sentence in dataframe
for text in combined_reviews['text']:
    nlp_doc = nlp(text)
    sentence_list_spacy = [sentence for sentence in nlp_doc.sents]
    token_list_spacy = [token for token in nlp_doc]

```

```python
from collections import Counter


# Load spaCy English tokenizer, tagger, parser, NER, and word vectors
nlp = spacy.load("en_core_web_sm")

# Function to extract nouns from a piece of text
def extract_nouns(text):
    doc = nlp(text)
    nouns = [token.text.lower() for token in doc if token.pos_ == 'NOUN']
    return nouns

# Filter the reviews by star rating
one_star_reviews = combined_reviews[combined_reviews['stars'] == 1]
five_star_reviews = combined_reviews[combined_reviews['stars'] == 5]

# Apply the function to the 'text' column and extract nouns for 1-star reviews
nouns_one_star = one_star_reviews['text'].apply(extract_nouns).explode()

# Apply the function to the 'text' column and extract nouns for 5-star reviews
nouns_five_star = five_star_reviews['text'].apply(extract_nouns).explode()

# Find the top 20 most common nouns in 1-star reviews
top20_nouns_one_star = Counter(nouns_one_star).most_common(20)

# Find the top 20 most common nouns in 5-star reviews
top20_nouns_five_star = Counter(nouns_five_star).most_common(20)

# Display the results for 1-star reviews
print("Top 20 nouns in 1-star reviews:")
for word, freq in top20_nouns_one_star:
    print(f"{word}: {freq}")
```

```python
#Analyzing Displaying top 20 nouns from five star reviews
nouns_five_star = five_star_reviews['text'].apply(extract_nouns).explode()
top20_nouns_five_star = Counter(nouns_five_star).most_common(20)
print("\nTop 20 nouns in 5-star reviews:")
for word, freq in top20_nouns_five_star:
  print(f"{word}: {freq}")
```

```python
# Analyzing and finding Top 20 adjectives and verbs of one star and five star reviews
# Load spaCy English tokenizer, tagger, parser, NER, and word vectors
nlp = spacy.load("en_core_web_sm")

# Function to extract adjectives from a piece of text
def extract_adjectives(text):
    doc = nlp(text)
    adjectives = [token.text.lower() for token in doc if token.pos_ == 'ADJ']
    return adjectives

# Function to extract verbs from a piece of text
def extract_verbs(text):
    doc = nlp(text)
    verbs = [token.text.lower() for token in doc if token.pos_ == 'VERB']
    return verbs

# Filter the reviews by star rating from the 'selected_reviews' DataFrame
one_star_reviews = combined_reviews[combined_reviews['stars'] == 1]
five_star_reviews = combined_reviews[combined_reviews['stars'] == 5]

# Apply the adjective extraction function to the 'text' column for 1-star reviews
adjectives_one_star = one_star_reviews['text'].apply(extract_adjectives).explode()
# Apply the verb extraction function to the 'text' column for 1-star reviews
verbs_one_star = one_star_reviews['text'].apply(extract_verbs).explode()

# Find the top 20 most common adjectives in 1-star reviews
top20_adjectives_one_star = Counter(adjectives_one_star).most_common(20)
# Find the top 20 most common verbs in 1-star reviews
top20_verbs_one_star = Counter(verbs_one_star).most_common(20)

# Apply the adjective extraction function to the 'text' column for 5-star reviews
adjectives_five_star = five_star_reviews['text'].apply(extract_adjectives).explode()
# Apply the verb extraction function to the 'text' column for 5-star reviews
verbs_five_star = five_star_reviews['text'].apply(extract_verbs).explode()

```

```python
# Analyzing and Finding the top 20 most common adjectives in 5-star reviews
top20_adjectives_five_star = Counter(adjectives_five_star).most_common(20)
# Find the top 20 most common verbs in 5-star reviews
top20_verbs_five_star = Counter(verbs_five_star).most_common(20)

# Display the results for adjectives
print("Top 20 adjectives in 1-star reviews:")
for word, freq in top20_adjectives_one_star:
    print(f"{word}: {freq}")

print("\nTop 20 adjectives in 5-star reviews:")
for word, freq in top20_adjectives_five_star:
    print(f"{word}: {freq}")

# Display the results for verbs
print("\nTop 20 verbs in 1-star reviews:")
for word, freq in top20_verbs_one_star:
    print(f"{word}: {freq}")

print("\nTop 20 verbs in 5-star reviews:")
for word, freq in top20_verbs_five_star:
    print(f"{word}: {freq}")
```

```python
#Finding Top 20 Named entities in Selected Reviews
# Load spaCy's English model with named entity recognition
nlp = spacy.load("en_core_web_sm")

# Function to extract named entities from a piece of text
def extract_named_entities(text):
    doc = nlp(text)
    # Extract entities, filter out non-GPE (Geo-Political Entities) or PERSON if you want specific types
    entities = [ent.text for ent in doc.ents]
    return entities

# Apply the named entity extraction function to the 'text' column in selected_reviews
named_entities = combined_reviews['text'].apply(extract_named_entities).explode()

# Find the top 20 most common named entities in the selected reviews
top20_named_entities = Counter(named_entities).most_common(20)

# Display the results
print("Top 20 named entities in selected reviews:")
for entity, freq in top20_named_entities:
    print(f"{entity}: {freq}")
```
